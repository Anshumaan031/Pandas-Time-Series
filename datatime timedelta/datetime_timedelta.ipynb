```python
import pandas as pd
import datetime
```

The timedelta is a class in datetime module that represents duration. The delta means average of difference and so the duration expresses the difference between two date, datetime or time instances. By using timedelta, you may estimate the time for future and past

```python
date_base=datetime.datetime.today()             #for printing current date
Previous_Date = datetime.datetime.today() - datetime.timedelta(days=20)  #for priniting the previous 20 days
rng=pd.date_range(start=Previous_Date,end=date_base,freq='D')
```

```python
df=pd.DataFrame(rng)
df
```

```python
date_base=datetime.datetime.today()             
df_date_list=[date_base-datetime.timedelta(days=x) for x in range(0,20)]
df2=pd.DataFrame(df_date_list)
df2.columns=['day']
df2
```
