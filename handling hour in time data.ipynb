```python
drop=df[['state','county','community','communityname','fold']]
df=df.drop(drop, axis=1)
df.head()
```

```python
cols=df.columns
```

```python
#df.dtypes
```

```python
from sklearn.impute import SimpleImputer
mean_imp = SimpleImputer(missing_values='?', strategy='most_frequent')
df=mean_imp.fit_transform(df)
df
```

```python
df=pd.DataFrame(df, columns=cols)
df
```

```python
X=df.drop('ViolentCrimesPerPop', axis=1)
X
```

```python
Y=df[['ViolentCrimesPerPop']]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size=0.3)
```

```python
X_train.shape
```

```python
Y_train.shape
```

# linear Rgression

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
```

```python
model1.fit(X_train,Y_train)
```

```python
y_pred=model1.predict(X_test)
```

```python
model1.score(X_test,Y_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
rmse = np.sqrt(mean_squared_error(Y_test,y_pred))
r2 = r2_score(Y_test,y_pred)
print('Root mean sqaure error:',rmse)
print('R-Sqaured error:',r2)
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
model2=LinearRegression()
pf=PolynomialFeatures(degree=2)
x_poly=pf.fit_transform(X_train)
model2.fit(x_poly,Y_train)
```

```python
x_test_poly=pf.fit_transform(X_test)
```

```python
y_poly_pred = model2.predict(x_test_poly)
```

```python
rmse = np.sqrt(mean_squared_error(Y_test,y_poly_pred))
r2 = r2_score(Y_test,y_poly_pred)
print('Root mean sqaure error:',rmse)
print('R-Sqaured error:',r2)
```
